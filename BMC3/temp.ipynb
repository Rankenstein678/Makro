```python
from scipy.stats import linregress
import matplotlib.pyplot as plt
import numpy as np

xs = np.linspace(0, 25, 1000)



Tgs = [63.1,67.3,68.6]
Heiz = [5,10,20]

fig,ax = plt.subplots()
ax.scatter(Heiz,Tgs)
l = linregress(Heiz,Tgs)
ax.plot(xs,l.slope*xs+l.intercept,c="r",label = f"Ausgleichsgerade mit y-Achsenabschnitt$={l.intercept}$ K" )
ax.set_xlabel("Heizrate [Km$^{-1}$]")
ax.set_ylabel("Glas√ºberganstemperatur $T_g$ [K]")
ax.legend()
fig.savefig("Tg.pdf")
```
